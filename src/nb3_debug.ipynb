```python
import os
from ctypes import CDLL
```

```python
# Pfad zu deiner GDAL DLL
gdal_path = r"C:\Users\daniil\AppData\Local\Programs\OSGeo4W\bin\gdal311.dll"

# Prüfen, ob die Datei existiert
print("Existiert die GDAL DLL?", os.path.isfile(gdal_path))

# Versuche, die DLL zu laden
try:
    CDLL(gdal_path)
    print("✅ GDAL DLL konnte geladen werden!")
except OSError as e:
    print("❌ Fehler beim Laden der GDAL DLL:", e)
```
