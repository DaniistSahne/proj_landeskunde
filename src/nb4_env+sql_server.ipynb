```python
import os

sql_files_path = r"C:\Users\daniil\Desktop\proj_landeskunde\src\parser"
sql_files = [f for f in os.listdir(sql_files_path) if f.endswith('.sql')]

for file in sql_files:
    print(file)
```

```python
# create a combined python file "all_parser.py" in the parser directory
files = [f for f in os.listdir(sql_files_path) if f.endswith('.py')]

first = 'generate_places_sql.py'
ordered = []
if first in files:
    ordered.append(first)
    files.remove(first)
files_sorted = ordered + sorted(files)

out_path = os.path.join(sql_files_path, 'all_parser.py')
with open(out_path, 'w', encoding='utf-8') as out:
    for fname in files_sorted:
        src = os.path.join(sql_files_path, fname)
        out.write(f'# --- Begin: {fname} ---\n')
        with open(src, 'r', encoding='utf-8') as f:
            out.write(f.read())
        out.write(f'\n# --- End: {fname} ---\n\n')

print(f'Created {out_path} containing {len(files_sorted)} files (first: {files_sorted[0] if files_sorted else "none"}).')
```
